# 2. Revisão — Lista simplesmente encadeada

Uma **lista simplesmente encadeada** é formada por nós.

Cada nó normalmente possui:

- um **valor**;
- uma referência para o **próximo nó**.

Uma representação simples é:

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | / ]
```

Nesse desenho:

- `head` aponta para o primeiro nó;
- cada nó aponta para o próximo;
- o último nó aponta para `null` (ou `/`, para representar ausência de próximo).

## Ideia central

Em uma lista simplesmente encadeada:

- andamos **para frente**;
- não existe, por padrão, um ponteiro para o nó anterior;
- isso influencia diretamente o custo e a facilidade das operações.

## 2.1 Operações típicas em lista simplesmente encadeada

As operações mais comuns são:

- inserir no início;
- inserir no fim;
- inserir após um nó;
- remover do início;
- remover do fim;
- remover um nó específico;
- buscar um valor;
- percorrer a lista inteira.

Vamos rever o raciocínio de cada uma.

## 2.2 Inserção no início

Suponha a lista:

```text
head
 ↓
[20 | • ] → [30 | / ]
```

Queremos inserir o valor `10` no início.

### Passos conceituais

1. Criamos um novo nó com valor `10`;
2. O `próximo` desse novo nó passa a apontar para o antigo primeiro nó;
3. `head` passa a apontar para o novo nó.

### Resultado

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | / ]
```

### Observação importante

Essa operação é muito eficiente porque não precisamos percorrer a lista.  
Em geral, ela pode ser feita em **O(1)**.

## 2.3 Inserção no fim

Agora suponha:

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | / ]
```

Queremos inserir `40` no final.

### Passos conceituais

1. Criamos o nó `[40 | / ]`;
2. Precisamos localizar o último nó atual (`30`);
3. Fazemos o `próximo` do último nó apontar para o novo nó.

### Resultado

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | • ] → [40 | / ]
```

### Ponto de atenção

Se a estrutura tiver apenas `head`, normalmente precisamos percorrer a lista até o final.  
Nesse caso, a operação tende a custar **O(n)**.

Se também houver um ponteiro `tail`, inserir no fim pode se tornar mais eficiente.

## 2.4 Remoção do início

Lista inicial:

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | / ]
```

Para remover o primeiro nó:

1. guardamos a referência do primeiro nó, se necessário;
2. fazemos `head` apontar para o segundo nó.

### Resultado

```text
head
 ↓
[20 | • ] → [30 | / ]
```

Essa operação também costuma ser **O(1)**.

## 2.5 Remoção do fim

Lista inicial:

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | / ]
```

Queremos remover o último nó (`30`).

### Problema conceitual

Em uma lista simplesmente encadeada, o nó `30` não tem referência para o nó anterior.  
Então, para desconectá-lo corretamente, precisamos encontrar o nó anterior (`20`).

### Passos

1. percorremos a lista até encontrar o penúltimo nó;
2. fazemos o `próximo` do penúltimo nó apontar para `null`.

### Resultado

```text
head
 ↓
[10 | • ] → [20 | / ]
```

### Conclusão

Remover do fim em lista simplesmente encadeada costuma exigir percurso.  
Por isso, costuma ter custo **O(n)**.

## 2.6 Busca e percurso

Para buscar um valor em lista simplesmente encadeada, normalmente começamos em `head` e seguimos nó a nó.

Exemplo:

```text
head
 ↓
[10 | • ] → [20 | • ] → [30 | / ]
```

Se queremos encontrar o valor `30`, visitamos:

- primeiro o nó `10`;
- depois o nó `20`;
- depois o nó `30`.

Esse tipo de acesso é **sequencial**.

## Comparação rápida com vetor

Em vetor, o acesso por índice é direto.

Em lista encadeada:

- não há acesso imediato ao elemento “da posição 2” ou “da posição 10”;
- é preciso caminhar pela estrutura.

Esse detalhe é importante porque mostra que cada estrutura favorece certos tipos de operação.

# 3. Revisão — Pilhas

A **pilha** é uma estrutura baseada na política:

> **LIFO — Last In, First Out**  
> O último a entrar é o primeiro a sair.

## Intuição

Pense em uma pilha de pratos:

- você adiciona pratos no topo;
- você remove pratos do topo.

## Operações típicas

- `push`: empilha um elemento;
- `pop`: desempilha o elemento do topo;
- `top` ou `peek`: consulta o topo sem remover;
- `is_empty`: verifica se a pilha está vazia.

## 3.1 Exemplo visual de pilha

Pilha inicialmente vazia:

```text
Topo
 ∅
```

Após `push(10)`:

```text
Topo
 ↓
[10]
```

Após `push(20)`:

```text
Topo
 ↓
[20]
[10]
```

Após `push(30)`:

```text
Topo
 ↓
[30]
[20]
[10]
```

Agora, se fizermos `pop()`, quem sai é `30`.

### Resultado

```text
Topo
 ↓
[20]
[10]
```

## Ideia central

Em pilha:

- inserimos no topo;
- removemos do topo;
- não removemos do “meio” da estrutura no uso clássico.

## 3.2 Pilha com lista simplesmente encadeada

Uma pilha combina muito bem com lista simplesmente encadeada.

Por quê?

Porque tanto `push` quanto `pop` podem ser feitos no início da lista:

- `push` no topo → inserir no início;
- `pop` no topo → remover do início.

Como essas operações são eficientes em lista simplesmente encadeada, a pilha fica muito natural nessa implementação.

## 3.3 Aplicações clássicas de pilha

Alguns exemplos:

### a) Desfazer ações (Undo)
Em editores de texto, a ação mais recente é a primeira que normalmente queremos desfazer.

### b) Chamadas de função
Em muitas linguagens, as chamadas são organizadas em pilha de execução.

### c) Verificação de parênteses e símbolos
Ao encontrar um símbolo de abertura, empilhamos.  
Ao encontrar um símbolo de fechamento, verificamos se ele combina com o topo.

### d) Histórico de navegação para “voltar”
Ao visitar páginas, é possível empilhar os estados anteriores.

# 4. Revisão — Filas

A **fila** é uma estrutura baseada na política:

> **FIFO — First In, First Out**  
> O primeiro a entrar é o primeiro a sair.

## Intuição

Pense em uma fila de atendimento:

- quem chega primeiro, é atendido primeiro.

## Operações típicas

- `enqueue`: insere no fim da fila;
- `dequeue`: remove do início;
- `front`: consulta o primeiro elemento;
- `is_empty`: verifica se a fila está vazia.

## 4.1 Exemplo visual de fila

Fila vazia:

```text
Início → ∅ ← Fim
```

Após `enqueue(10)`:

```text
Início → [10] ← Fim
```

Após `enqueue(20)`:

```text
Início → [10] → [20] ← Fim
```

Após `enqueue(30)`:

```text
Início → [10] → [20] → [30] ← Fim
```

Agora, se fizermos `dequeue()`, quem sai é `10`.

### Resultado

```text
Início → [20] → [30] ← Fim
```

## Ideia central

Na fila:

- entra pelo fim;
- sai pelo início.

## 4.2 Fila com lista simplesmente encadeada

Fila também pode ser implementada com lista simplesmente encadeada.

### Estratégia típica

- inserir no fim;
- remover do início.

### Observação importante

Para que a fila fique eficiente, costuma ser útil manter:

- `head` apontando para o início;
- `tail` apontando para o fim.

Assim:

- `enqueue` pode ser feito no fim de forma eficiente;
- `dequeue` pode ser feito no início de forma eficiente.

## 4.3 Aplicações clássicas de fila

### a) Atendimento
Ordem de chegada.

### b) Impressão
Documentos enviados primeiro tendem a ser processados antes.

### c) Processamento de tarefas
Em muitos cenários, tarefas são colocadas em fila para execução.

### d) Simulações
Filas aparecem em problemas de filas de banco, filas de rede, escalonamento simples etc.

# 5. Comparando lista, pilha e fila

Até aqui, é importante separar as ideias:

## Lista simplesmente encadeada
É uma **forma de organização** dos nós.

## Pilha
É um **TAD** com regra LIFO.

## Fila
É um **TAD** com regra FIFO.

Isso quer dizer que:

- pilha e fila podem ser implementadas usando diferentes estruturas internas;
- a lista encadeada é uma possível implementação.

> Em outras palavras:  
> **pilha e fila dizem “como usar”**;  
> **lista encadeada diz “como organizar internamente”.**

# 6. Aplicações e escolha da estrutura

Uma parte muito importante do estudo de Estruturas de Dados é responder:

> **Por que escolhi essa estrutura e não outra?**

Vamos pensar em alguns cenários.

## 6.1 Cenário 1 — Desfazer ações em um editor

Quando o usuário digita e depois quer desfazer:

- a ação mais recente deve ser revertida primeiro.

### Estrutura adequada
**Pilha**

### Justificativa
A operação segue a lógica de “último que entrou, primeiro que sai”.

## 6.2 Cenário 2 — Fila de impressão

Vários documentos chegam para impressão.  
Em uma política simples, queremos processar por ordem de chegada.

### Estrutura adequada
**Fila**

### Justificativa
A ordem natural é “primeiro que entra, primeiro que sai”.

## 6.3 Cenário 3 — Verificação de expressões com parênteses

Exemplo:

```text
(a + b) * (c - d)
```

Ao encontrar `(`, empilhamos.  
Ao encontrar `)`, verificamos o topo.

### Estrutura adequada
**Pilha**

### Justificativa
Precisamos combinar o símbolo de fechamento com o símbolo de abertura mais recente ainda não fechado.

## 6.4 Cenário 4 — Atendimento comum

Cada pessoa chega e entra no final.  
O atendimento chama quem está no começo.

### Estrutura adequada
**Fila**

### Justificativa
O atendimento respeita a ordem de chegada.

# 7. O problema: pilha e fila não resolvem tudo

Agora chegamos ao ponto central da aula.

Pilha e fila são muito úteis, mas existem problemas em que precisamos operar **nas duas extremidades** da estrutura.

## Exemplo de necessidade

Imagine uma sequência em que precisamos:

- inserir elementos no início;
- inserir elementos no fim;
- remover elementos do início;
- remover elementos do fim.

Vamos pensar:

### Pilha resolve?
Não, porque ela centraliza as operações no topo.

### Fila resolve?
Não totalmente, porque a fila clássica favorece inserção no fim e remoção no início.

Então surge a necessidade de uma estrutura mais flexível.

# 8. Introdução ao deque

**Deque** vem de **double-ended queue**, isto é, uma fila com duas extremidades operacionais.

## Ideia central

No deque, podemos operar dos dois lados.

### Operações típicas

- `push_front`: inserir no início;
- `push_back`: inserir no fim;
- `pop_front`: remover do início;
- `pop_back`: remover do fim;
- `front`: consultar o início;
- `back`: consultar o fim.

## Representação

```text
Início ⇄ [10] ⇄ [20] ⇄ [30] ⇄ Fim
```

A grande ideia é:

> o deque é útil quando a aplicação precisa de flexibilidade nas extremidades.

## 8.1 Deque é uma estrutura ou uma implementação?

Essa distinção é muito importante.

**Deque é um TAD**, isto é, um conjunto de operações e comportamentos esperados.

Ele pode ser implementado de diferentes formas, por exemplo:

- vetor circular;
- lista duplamente encadeada;
- em alguns casos, lista simplesmente encadeada (com limitações).

## Implementação mais natural

Em muitos contextos, a implementação mais natural do deque é com **lista duplamente encadeada**, porque ela facilita operações eficientes nas duas extremidades.

## 8.2 Faz sentido pensar deque com lista simplesmente encadeada?

Sim, faz sentido **didaticamente**.

### Por que faz sentido em aula?

Porque ajuda o aluno a perceber:

- o que a estrutura precisa fazer;
- quais operações são fáceis;
- quais operações se tornam difíceis;
- por que uma lista duplamente encadeada pode ser necessária.

### Limitação importante

Em lista simplesmente encadeada, remover do fim geralmente exige percurso.

Então, em um deque implementado com lista simples:

- `push_front` tende a ser eficiente;
- `pop_front` tende a ser eficiente;
- `push_back` pode ser eficiente se houver `tail`;
- `pop_back` tende a ser mais caro, porque precisamos achar o penúltimo nó.

Isso é excelente do ponto de vista pedagógico, porque mostra que:
> **a escolha da implementação afeta o custo das operações.**

# 10. Onde o deque pode ser usado?

Alguns exemplos:

## a) Janela deslizante
Problemas em que mantemos elementos recentes na estrutura e descartamos antigos por uma das pontas.

## b) Histórico com navegação em duas extremidades
Em alguns cenários, precisamos acrescentar ou retirar elementos dos dois lados.

## c) Algoritmos específicos
Existem problemas clássicos em que o deque ajuda a manter candidatos, máximos, mínimos ou estados relevantes nas extremidades.

## d) Modelagem flexível
Quando o sistema permite comportamentos híbridos que não se encaixam exatamente nem em pilha nem em fila.
