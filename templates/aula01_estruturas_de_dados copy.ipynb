**Conceito**

- append vs insert
- custo intuitivo
- estrutura vs implementação


```python
# ------------------------------------------------------------
# Comparando operações
# ------------------------------------------------------------

import time

lista = []

# append
start = time.time()
for i in range(100000):
    lista.append(i)
end = time.time()

print("Tempo append:", end - start)

# insert no início
lista = []
start = time.time()
for i in range(10000):
    lista.insert(0, i)
end = time.time()

print("Tempo insert:", end - start)
```
