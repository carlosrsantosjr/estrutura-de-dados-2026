# Listas Simplesmente Encadeadas

Operações para usar uma lista encadeada em aplicações de tempo real:

* Percorrer a lista
* Inserir um item de dados na lista:
    * No começo
    * No fim
    * No meio ou qualquer outra posição
* Excluir um item da lista
    * Primeiro Nó
    * Último Nó
    * Meio ou qualquer outra posição


```python
class Node:
    def __init__(self, data=None):
        self.data  = data
        self.next = None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next
    
    # Inserção no fim (s/ tail)
    '''
    def append(self, data):
        if self.head:
            pointer = self.head
            while pointer.next:
                pointer = pointer.next
            pointer.next = Node(data)
        else:
            self.head = Node(data)
        self.size = self.size + 1
    '''

    # Inserção no fim (c/ tail)
    def append(self, data):
        node = Node(data)
        if self.tail:
            self.tail.next = node
            self.tail = node
        else:
            self.head = node
            self.tail = node
        self.size = self.size + 1

    # Inserção no meio
    def append_at_a_location(self, data, index):
        node_new = Node(data)
        pointer = self.head
        for i in range(index):
            if pointer:
                pointer_bef = pointer
                pointer = pointer.next
            else:
                raise IndexError("list index out of range")
        if pointer:
            pointer_bef.next = node_new
            node_new.next = pointer
        else:
            raise IndexError("list index out of range")

    def insert_0(self, data, index):
        node_new = Node(data)
        pointer = self.head
        pointer_bef = self.head
        
        # Pecorre a lista e posiciona o ponteiro no index
        for i in range(index):          
            print("1 index: ", i)
            if pointer:                 # Se o elemento existir.
                print("2 index: ", i)
                pointer_bef = pointer   # Guarda endereço do elemento anterior
                pointer = pointer.next  # Atualiza o ponteiro com o elemento posterior
            else:
                # lista pode estar vazia
                # lista pode ter acabado
                raise IndexError("list index out of range - 0001")
        # Insere o elemento no início da lista    
        if index == 0:                  
            node_new.next = self.head
            self.head = node_new
        # Insere o elemento no meio da lista
        elif pointer:                   
            pointer_bef.next = node_new # O elemento anterior aponta para o novo Nó
            node_new.next = pointer     # O novo nó aponta para o elemento atual do index
        # Insere o elemento no final da lista
        else:                               
            pointer_bef.next = node_new # O elemento anterior aponta para o novo Nó
```

```python
lista = SinglyLinkedList()
lista.insert_0(100,0)
lista.show()
```
