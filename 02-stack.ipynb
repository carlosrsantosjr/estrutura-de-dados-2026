```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.topo = None
        self._size = 0

    def push(self, data):
        new_node = Node(data)       # instancio um novo Nó
        new_node.next = self.topo   # 
        self.topo = new_node        
        self._size = self._size + 1

    def pop(self):
        if self.topo is None:
            return None
        removed = self.topo
        self.topo = self.topo.next
        self._size = self._size - 1
        return removed.data

    def peek(self):
        if self.topo is None:
            return None
        return self.topo.data
    
    def __len__(self):
        return self._size
```

```python
s = Stack()
s.push(10)
s.push(20)
print(s.peek())
s.push(30)
print(s.peek())
s.pop()
print(s.peek())
```
