```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
```

```python
no1 = Node(7)
no2 = Node(5)
no3 = Node(8)

no1.next = no2
no2.next = no3
```

```python
class LinkedList:
    
    # head aponta para o endereço do primeiro elemento da lista
    def __init__(self):
        self.head = None # Lista vazia
        self._size = 0

    def append(self, elem):             # Inserção de um elemento no final da lista
        if self.head:                   # Verifica se existe pelo menos um elemento
            pointer = self.head         # Variável auxiliar
            while(pointer.next):        # Avança até o último elemento (enquanto exisitir elementos)
                pointer = pointer.next
            pointer.next = Node(elem)   # Insere novo elmento no final da lista 
        else:
            self.head = Node(elem)                  # Insere primeiro elemento da
        self._size = self._size + 1
    
    def __len__(self):                  # Retorna o tamanho da lista
        return self._size
    
    def __getitem__(self, index):
        pointer = self.head             # Variável auxiliar
        for i in range(index):          # Percorre a lista
            if pointer:                 # Se lista possui pelo menos um elemento
                pointer = pointer.next
            else:
                raise IndexError("list index out of range")
        if pointer:
            return pointer.data
        else:
            raise IndexError("list index out of range")
    
    def __setitem__(self, index, elem):
        pointer = self.head             # Variável auxiliar
        for i in range(index):          # Percorre a lista
            if pointer:                 # Se lista possui pelo menos um elemento
                pointer = pointer.next
            else:
                raise IndexError("list index out of range")
        if pointer:
            pointer.data = elem
        else:
            raise IndexError("list index out of range")
```

```python
lista = LinkedList()
lista.append(5)
lista.append(5)
```
