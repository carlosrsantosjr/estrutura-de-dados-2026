## Tipos Abstratos de Dados (TADs)

À medida que a complexidade dos sistemas cresce, torna-se necessário a aplicação do conceito de encapsulamento. Por exemplo, uma vez definida uma função, toda a lógica interna fica encapsulada do usuário. Sempre que necessário, o programador apenas invoca a função, sem precisar conhecer os detalhes de sua implementação.

O mesmo princípio não se aplica apenas a processos, mas também aos dados. É nesse contexto que surgem os **Tipos Abstratos de Dados (TAD’s)**, que permitem estruturar e organizar informações de maneira controlada e modular.

Um Tipo Abstrato de Dados (TAD) é um tipo definido pelo programador que especifica:

- Um conjunto de variáveis responsáveis por armazenar informações;
- Um conjunto de operações bem definidas que podem ser aplicadas sobre essas variáveis.

A principal característica de um TAD é o encapsulamento. Sua implementação interna é ocultada, de modo que o acesso aos dados ocorre exclusivamente por meio de uma interface, definida pelo conjunto de operações disponíveis.

Assim, o usuário do TAD interage apenas com essa interface, sem necessidade de conhecer como os dados são armazenados ou como as operações são implementadas internamente. Essa separação entre interface e implementação favorece modularidade, reutilização de código e manutenção de sistemas.
A grande vantagem desse modelo é que, uma vez definido o TAD, sua utilização depende apenas do conhecimento das operações oferecidas pela interface, independentemente dos detalhes internos de implementação.

Alguns exemplos de TADs são:

- TAD Lista (Sequencial)
- TAD Fila
- TAD Pilha
- TAD Árvore

## Listas Encadeadas

As estruturas de dados comumente utilizadas até aqui são conhecidas como **estruturas lineares estáticas**, implementadas tipicamente por
meio de vetores (arrays). Essas estruturas têm as seguites caracteristicas:

- A quantidade máxima de elementos deve ser previamente definida;
- O armazenamento ocorre de forma contígua na memória. 

Embora essa abordagem seja simples e eficiente para diversas aplicações, ela apresenta limitações importantes, especialmente quando:

- O número de elementos não é conhecido antecipadamente; 
- Variação significativa ao longo da execução do programa.

Isso acontece devido as frequentes movimentações de memória quando acionadas as aperações de **inserção** ou **remoção**, cuja a complexidade é O(n).

Afim de superar essas limitações existem as **estruturas de dados lineares dinâmicas**. Diferentemente dos vetores tradicionais, nessas estruturas os elementos não precisam ocupar posições contíguas de memória. Cada elemento é alocado individualmente e conectado aos demais por meio
de referências (ou ponteiros), formando um encadeamento lógico. Essa característica permite que a estrutura cresça ou diminua conforme necessário, utilizando memória de maneira mais flexível.

Entre as principais estruturas lineares dinâmicas destacam-se:
- Listas Simplesmente Encadeadas;
- Listas Duplamente Encadeadas.

O estudo dessas estruturas é fundamental para compreender conceitos centrais como alocação dinâmica de memória, manipulação de ponteiros e custo de operações de inserção e remoção. Além disso, listas encadeadas servem como base para a implementação de diversas outras estruturas mais complexas, como pilhas, filas, tabelas de dispersão e árvores.

### Listas Simplesmente Encadeadas

Em programação de computadores, é possível utilizar estruturas de dados de maneira dinâmica, isto é, projetar estruturas que **podem crescer ou diminuir de tamanho em tempo de execução**, seja por meio da inserção ou da remoção de elementos. Essa característica é especialmente importante em
situações nas quais o número de dados a ser armazenado não é conhecido previamente ou pode variar significativamente ao longo da execução do programa.

Diferentemente das estruturas baseadas em vetores estáticos, nas quais o espaço deve ser previamente reservado de forma contígua, as estruturas dinâmicas exploram o conceito de referências (ou ponteiros) para conectar elementos armazenados em posições arbitrárias da memória.

Nesse contexto, as listas encadeadas constituem um exemplo clássico de estrutura dinâmica. Elas utilizam um esquema de encadeamentos lógicos sequenciais, no qual:
- **cada elemento (nó) mantém uma referência para o próximo elemento da sequência**.

Dessa forma, elementos vizinhos na estrutura lógica não precisam ocupar posições contíguas na memória física, como ilustrado na figura a seguir.

#### Nós e Ponteiros

Primeiramente, antes de definirmos formalmente uma lista encadeada, é necessário especificar seu bloco fundamental de construção: o nó. Cada nó de uma lista encadeada deve conter duas informações essenciais: 
- o dado propriamente dito e 
- uma referência (ponteiro) para o próximo nó da lista

Podemos definir um TAD Node, que representa um nó da lista, da seguinte forma:

```python
class Node:
    def __init__(self, data=None):
        self.data = data            # armazena o dado
        self.next = None            # referência para o próximo nó 
```



Figura 3: Ilustração da estrutura básica Node, que define um nó em uma lista simplesmente encadeada.


#### Criando e Percorrendo uma Lista Simplesmente Encadeada

Instanciaremos seis objetos da classe Nó adicionando os elementos apresentados na Figura 2. 

```python
# criação dos Nós - Instanciação dos objetos da classe Node
no1 = Node(54)
no2 = Node(26)
no3 = Node(93)
no4 = Node(17)
no5 = Node(77)
no6 = Node(13)
```

Em seguida, vincularemos os nós sequencialmente para formar a lista encadeada. Veja que o *no1* está apontando para o *no2*, o *no2* está apontando para o *no3*, e seguem dessa forma até o último *no6* que aponta para *None*, conforme definido na classe.

```python
# vinculação dos nós através do atributo next
no1.next = no2
no2.next = no3
no3.next = no4
no4.next = no5
no5.next = no6
```

AQUIIII

```python
pointer = no1
while pointer:
    print(pointer.data)
    pointer = pointer.next
```

#### Lista Simplesmente Encadeada - Implementação

Em uma lista encadeada, a principal característica é que:

- **a inserção de novos nós é realizada tipicamente no início (ou eventualmente no final) da lista**

Inicialmente, apresentaremos a criação de uma lista encadeada vazia. Adotaremos o seguinte TAD, no qual head (cabeça) é uma referência para o primeiro nó da lista:

```python
# Classe 
class LinkList:

    def __init__(self):
        self.head = None
        self._size = 0
```



Figura 4: Ilustração da inserção de um novo nó no final de uma lista simplesmente encadeada.



Figura 5: Ilustração da busca por um nó em uma lista simplesmente encadeada.

```python
class LinkList:

    def __init__(self):
        self.head = None
        self._size = 0

    def append(self, elem):
        if self.head:                   # Tem pelo menos um elemento?
            pointer = self.head         
            while(pointer.next):        # Percorrer a lista até next.None
                pointer = pointer.next
            pointer.next = Node(elem)   # Adiciona um novo elemento
        else:
            self.head = Node(elem)      # Adiciona o primeiro elemento
        self._size = self._size + 1

    def __len__(self):
        return self._size
    
    def __getitem__(self, index):
        pointer = self.head
        for i in range(index):
            if pointer:
                pointer = pointer.next
            else:
                raise IndexError("list index out of range")   
        if pointer:
            return pointer.data
        else:
            raise IndexError("list index out of range") 
        
    def __setitem__(self, index, elem):
        pointer = self.head
        for i in range(index):
            if pointer:
                pointer = pointer.next
            else:
                raise IndexError("list index out of range")   
        if pointer:
            pointer.data = elem
        else:
            raise IndexError("list index out of range") 
```

```python
lista = LinkList()
lista.append(5)
lista.append(7)
lista.append(9)
print(lista[1])
lista[1] = 20
print(lista[1])
```
