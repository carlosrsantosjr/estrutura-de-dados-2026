# Deque: Double-Ended Queue (Fila de Duas Extremidades)

Deque (pronuncia-se deck, para diferenciar da operação dequeue), ou Double-Ended Queue, que traduziremos como Dupla Fila, é uma estrutura de dados linear bastante similar a uma fila tradicional. O que a torna diferente da fila convencional é justamente a possibilidade de inserir e remover elementos tanto do início quanto do fim da estrutura. Nesse sentido, esse TAD híbrido **provê todas as funcionalidades de filas e pilhas** em uma única estrutura de dados. Para fixar a notação, denotaremos por *start* a parte da frente da fila (lado esquerdo) e por *end* a parte de trás da fila (lado direito), conforme apresentado a seguir:

*start* → [1, 2, 3, 4, 5] ← *end*

As principais operações de um Deque são listadas a seguir.

- *is_empty()*: verifica se o deque está vazio.
- *add_start(x)*: insere o elemento 𝑥no início.
- *add_end(x)*: insere o elemento 𝑥no final.
- *remove_start()*: remove e retorna o elemento do início.
- *remove_end()*: remove e retorna o elemento do final.
- *size()*: retorna o número de elementos armazenados.

### Aplicação do Deque: Verificação de Palíndromos

Um exemplo clássico de aplicação de uma estrutura linear do tipo Deque é o problema de verificar
se uma dada palavra é palíndroma, isto é, se ela é igual à sua versão reversa. Formalmente, uma
palavra é dita palíndroma quando sua sequência de caracteres lida da esquerda para a direita é idêntica à sequência lida da direita para a esquerda. Um exemplo simples é a palavra **RADAR**.
A ideia para resolver esse problema utilizando um Deque consiste em armazenar os caracteres da
string de entrada e, em seguida, comparar simetricamente os elementos das duas extremidades. O
procedimento pode ser descrito da seguinte forma:
1. Percorrer a string de entrada da esquerda para a direita, inserindo cada caractere no final do deque.
2. Enquanto houver mais de um elemento armazenado, remover simultaneamente um caractere do início (front) e outro do final (rear).
3. Comparar os dois caracteres removidos
- Se forem diferentes, a palavra não é palíndroma.
- Se forem iguais, o processo continua.
4. O algoritmo termina quando:
- o deque ficar vazio (caso de palavra com número par de caracteres), ou
- restar apenas um caractere (caso de palavra com número ímpar de caracteres).

```python
class Node:
    def __init__(self, data=None):
        self.data = data            # armazena o dado
        self.next = None            # referência para o próximo nó  

class Deque:
    def __init__(self):
        self.head = None            # indica a posição para remoção (início da fila);
        self.tail = None            # indica a posição para inserção (final da fila);
        self._size = 0              # armazena o número atual de elementos na fila.

    # Insere um elemento no início
    def add_start(self, elem):
        node = Node(elem)
        if self.head is None:           # Vazia
            self.tail = node            # Último nó da lista aponta para o novo nó 
            self.head = node            # Primeiro nó da lista aponta para o novo nó             
        else:
            # Novo elemento aponta para o primeiro nó
            # Novo nó passa a ser o primeiro nó
        self._size = self._size + 1


    # Insere um elemento no final
    def add_end(self, elem):
        node = Node(elem)
        if self.head is None:           # Vazia
            self.tail = node            # Último nó da lista aponta para o novo nó 
            self.head = node            # Primeiro nó da lista aponta para o novo nó             
        else:
            # Novo elemento aponta para o primeiro nó
            # Novo nó passa a ser o primeiro nó
        self._size = self._size + 1

    # Remove um elemento do início
    def remove_start(self):
        if self._size > 0:
            #
            #
            self._size = self._size - 1
            return elem
        raise IndexError("The queue is empty")
    
    # Remove um elemento no final
    def remove_end(self):
        if self._size > 0:
            
            self._size = self._size - 1
            return elem
        raise IndexError("The queue is empty")    

    # retorna o tamanho da fila
    def size(self):
        return self._size
    
    # retorna itens da fila
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next

    def is_empty(self):
        if self._size == 0:
            return True
        return False
        
```
