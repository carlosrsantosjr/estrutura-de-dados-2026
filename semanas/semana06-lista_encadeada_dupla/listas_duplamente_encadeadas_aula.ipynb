Até aqui, já trabalhamos com:

- **listas simplesmente encadeadas**;
- **pilhas**;
- **filas**.

Essas estruturas ajudam muito a compreender o encadeamento entre nós e a ideia de acesso sequencial.

Agora avançamos para uma estrutura em que cada nó passa a conhecer:

- quem está **à frente**;
- e também quem está **atrás**.

Quando estudamos listas simplesmente encadeadas, vimos que cada nó possui apenas uma referência para o próximo elemento.

Representação simplificada:

[dado | next] -> [dado | next] -> [dado | next]

Isso funciona bem em muitos cenários. Porém, há limitações importantes:

- só conseguimos caminhar naturalmente em **uma direção**;
- remover um nó interno pode exigir conhecer o **nó anterior**;
- certas operações ficam mais trabalhosas;
- navegar para trás não é natural.

Em vários problemas reais, isso é inconveniente.

Exemplos:

- histórico de navegação;
- sistemas de desfazer/refazer;
- caches com atualização frequente de posição;
- estruturas em que precisamos inserir ou remover elementos nas duas extremidades;
- cenários em que já temos a referência de um nó interno e queremos removê-lo rapidamente.

É nesse contexto que entram as **listas duplamente encadeadas**.

# Ideia central da lista duplamente encadeada

Em uma lista duplamente encadeada, cada nó possui:

- o valor armazenado;
- uma referência para o **próximo** nó;
- uma referência para o **nó anterior**.

Representação:

`None <- [prev | dado | next] <-> [prev | dado | next] <-> [prev | dado | next] -> None`

Ou seja:

- o primeiro nó tem `prev = None`;
- o último nó tem `next = None`.

Essa estrutura permite caminhar:

- da esquerda para a direita;
- da direita para a esquerda.

Também facilita ajustes locais quando já temos o nó que será manipulado.

# Comparação inicial com as estruturas já estudadas

## Lista simplesmente encadeada
- cada nó conhece apenas o próximo;
- navegação natural em uma direção;
- remoção de um nó geralmente exige conhecer o anterior.

## Pilha
- operações concentradas em uma extremidade;
- ótima quando a regra é **LIFO**.

## Fila
- inserção em uma ponta e remoção na outra;
- ótima quando a regra é **FIFO**.

## Lista duplamente encadeada
- maior flexibilidade estrutural;
- cada nó conhece o anterior e o próximo;
- permite navegação bidirecional;
- facilita remoções e inserções locais;
- exige mais memória por nó.

# Quando essa estrutura faz sentido?

A lista duplamente encadeada faz sentido quando:

- precisamos navegar nos dois sentidos;
- precisamos remover nós internos com frequência;
- já temos, em certos algoritmos, a referência do nó que será removido;
- queremos implementar estruturas mais flexíveis, como um **deque** baseado em nós;
- o custo extra de memória é aceitável.

Ela **não** é a melhor escolha quando:

- o problema exige acesso aleatório frequente por índice;
- o uso de memória precisa ser muito econômico;
- uma pilha ou fila já resolve o problema de forma mais simples.

# Estrutura do nó

Vamos começar pela classe do nó.

Cada nó terá:

- `data`: valor armazenado;
- `prev`: referência para o nó anterior;
- `next`: referência para o próximo nó.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
```

# Estrutura da lista

Na classe da lista, vamos manter:

- `head`: primeiro nó;
- `tail`: último nó;
- `_size`: quantidade de elementos.

Ter a referência para `tail` é importante porque:

- facilita inserções no final;
- facilita percursos invertidos;
- torna algumas operações mais elegantes.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next
```

# Inserção no início

### Caso 1: lista vazia
O novo nó será ao mesmo tempo:

- `head`
- `tail`

### Caso 2: lista com elementos
O novo nó deve:

- 1. apontar para o antigo `head`;
- 2. o antigo `head` deve apontar para trás para o novo nó;
- 3. `head` passa a ser o novo nó.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next

    # Inserção no Início
    def add_first(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

        self._size += 1
```

```python
dll = DoublyLinkedList()
dll.add_first(30)
dll.add_first(20)
dll.add_first(10)
dll.show()
```

# Inserção no final

### Caso 1: lista vazia
O novo nó será `head` e `tail`.

### Caso 2: lista não vazia
O novo nó:
- 1. terá `prev` apontando para o antigo `tail`;
- 2. o antigo `tail` terá `next` apontando para o novo nó;
- 3. `tail` passa a ser o novo nó.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next
    
    # Inserção no Início
    def add_first(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

        self._size += 1

    # Inserção no Final
    def add_last(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

        self._size += 1
```

```python
dll = DoublyLinkedList()
dll.add_last(10)
dll.add_last(20)
dll.add_last(30)
dll.show()
```

# Remoção no início

### Caso 1: lista vazia
Não há elemento para remover.

### Caso 2: lista com um único elemento
Após a remoção:
- `head = None`
- `tail = None`

### Caso 3: lista com mais de um elemento
O segundo nó passa a ser o novo `head`, e seu `prev` deve virar `None`.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next

    # Inserção no Início
    def add_first(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

        self._size += 1

    # Inserção no Final
    def add_last(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

        self._size += 1

    # Remoção no Início
    def remove_first(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.head.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None

        self._size -= 1
        return elem
```

```python
dll = DoublyLinkedList()
dll.add_last(10)
dll.add_last(20)
dll.add_last(30)

print("Antes:")
dll.show()
removed = dll.remove_first()
print("Removido:", removed)
print("Depois:")
dll.show()
```

# Remoção no final

### Caso 1: lista vazia
Erro.

### Caso 2: lista com um elemento
A lista volta a ficar vazia.

### Caso 3: lista com vários elementos
O penúltimo nó vira o novo `tail`, e seu `next` deve virar `None`.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0
    
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next

    # Inserção no Início
    def add_first(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

        self._size += 1

    # Inserção no Final
    def add_last(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

        self._size += 1

    # Remoção no Início
    def remove_first(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.head.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None

        self._size -= 1
        return elem

    # Remoção no Final
    def remove_last(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.tail.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None

        self._size -= 1
        return elem
```

```python
dll = DoublyLinkedList()
dll.add_last(10)
dll.add_last(20)
dll.add_last(30)

print("Antes:")
dll.show()
removed = dll.remove_last()
print("Removido:", removed)
print("Depois:")
dll.show()
```

# Remoção de um nó interno quando já temos o nó

Agora chegamos ao ponto mais interessante da estrutura.

Suponha que já temos a referência do nó que será removido.

Esse é justamente o tipo de cenário em que a lista duplamente encadeada se destaca.

## Ideia

Se o nó está no meio, basta ajustar:

- o `next` do nó anterior;
- o `prev` do nó seguinte.

Vamos implementar um método que recebe o **próprio nó**.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0
    
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next

    # Inserção no Início
    def add_first(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

        self._size += 1

    # Inserção no Final
    def add_last(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

        self._size += 1

    # Remoção no Início
    def remove_first(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.head.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None

        self._size -= 1
        return elem

    # Remoção no Final
    def remove_last(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.tail.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None

        self._size -= 1
        return elem

    def search(self, value):
        current = self.head
        while current:
            if current.data == value:
                return current
            current = current.next
        return None

    # Remoção no meio
    def remove_node(self, node):
        if node is None:
            raise ValueError("Invalid node")

        if node == self.head:
            return self.remove_first()

        if node == self.tail:
            return self.remove_last()

        node.prev.next = node.next
        node.next.prev = node.prev
        self._size -= 1
        return node.data
```

```python
dll = DoublyLinkedList()
dll.add_last(10)
dll.add_last(20)
dll.add_last(30)

print("Antes:")
dll.show()
node_30 = dll.search(20)
removed = dll.remove_node(node_30)
print("Removido:", removed)
print("Depois:")
dll.show()
```

## Complexidade dessa remoção

### Se o nó já foi dado
**`O(1)`**

### Se antes for necessário buscar o nó
- busca: `O(n)`
- remoção local: `O(1)`

**Total:** `O(n)`

Esse ponto é muito importante para comparação com a lista simplesmente encadeada.

Na lista simplesmente encadeada, mesmo tendo o nó, ainda pode ser necessário redescobrir o anterior.

# Inserção após um nó específico

Outro caso interessante é inserir depois de um nó já conhecido.

Isso também mostra bem a força da estrutura quando já temos um ponto interno.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0
    
    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0
    
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next

    # Inserção no Início
    def add_first(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

        self._size += 1

    # Inserção no Final
    def add_last(self, elem):
        new_node = Node(elem)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

        self._size += 1

    # Remoção no Início
    def remove_first(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.head.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None

        self._size -= 1
        return elem

    # Remoção no Final
    def remove_last(self):
        if self.is_empty():
            raise IndexError("The list is empty")

        elem = self.tail.data

        if self._size == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None

        self._size -= 1
        return elem

    def search(self, value):
        current = self.head
        while current:
            if current.data == value:
                return current
            current = current.next
        return None

    # Remoção no meio
    def remove_node(self, node):
        if node is None:
            raise ValueError("Invalid node")

        if node == self.head:
            return self.remove_first()

        if node == self.tail:
            return self.remove_last()

        node.prev.next = node.next
        node.next.prev = node.prev
        self._size -= 1
        return node.data
    
    def insert_after(self, node, elem):
        if node is None:
            raise ValueError("Invalid node")

        if node == self.tail:
            self.add_last(elem)
            return

        new_node = Node(elem)
        new_node.prev = node
        new_node.next = node.next

        node.next.prev = new_node
        node.next = new_node

        self._size += 1
```

```python
dll = DoublyLinkedList()
dll.add_last(10)
dll.add_last(20)
dll.add_last(40)

print("Antes:")
dll.show()
node_20 = dll.search(20)
dll.insert_after(node_20, 30)
print("Depois:")
dll.show()
```

## Complexidade

Se já temos o nó: **`O(1)`**

Se for preciso buscá-lo antes: **`O(n)`**

# Comparação didática: lista simples x lista dupla

## Lista simplesmente encadeada
### Vantagens
- implementação mais simples;
- menor custo de memória por nó;
- suficiente para pilhas, filas simples e muitos percursos lineares.

### Desvantagens
- não navega para trás;
- remoção no final costuma ser pior;
- remoção de nó interno é menos conveniente;
- algumas operações exigem guardar o nó anterior externamente.

## Lista duplamente encadeada
### Vantagens
- navegação bidirecional;
- remoção no final eficiente;
- remoção de nó interno mais elegante quando o nó é conhecido;
- inserções locais mais naturais.

### Desvantagens
- mais memória por nó;
- mais ponteiros para manter corretamente;
- maior chance de erros de implementação.
