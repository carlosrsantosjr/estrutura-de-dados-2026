# Filas

Assim como no caso das pilhas, é possível implementar uma fila de maneira dinâmica utilizando listas encadeadas. Para isso, definimos o Tipo Abstrato de Dados Queue.

```python
class Node:
    def __init__(self, data=None):
        self.data = data            # armazena o dado
        self.next = None            # referência para o próximo nó 
```

```python
class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
        self._size = 0
```

Durante a criação da fila, devemos inicializar a referência *head* com *None*, indicando que a fila está vazia, e definir o tamanho como zero.

A operação *enqueue()* insere um novo elemento no final do encadeamento, isto é, na cauda da fila.
Observe que, nessa implementação, a operação *enqueue()* possui complexidade *𝑂(1)*, pois envolve apenas a atualização da referência *tail*. Essa referência elimina a necessidade de percorrer toda a lista até alcançar o último nó para inserir o novo elemento.

```python
# Insere um elemento na fila
def enqueue(self, elem):
        node = Node(elem)
        if self.tail is None:           # Fila vazia
            self.tail = node            # Último nó da lista aponta para o novo nó 
            self.head = node            # Primeiro nó da lista aponta para o novo nó             
        else:
            self.tail.next = node       # Último elemento aponta para o novo nó
            self.tail = node            # Novo nó passa a ser o último elemento da fila
        self._size = self._size + 1
```

A operação *dequeue()* remove o elemento presente no início da fila:

```python
# Remove um elemento do ínicio da fila
def unqueue(self):
    if self._size > 0:
        elem  = self.head.data
        self.first = self.head.next
        self._size = self._size - 1
        return elem
    raise IndexError("The queue is empty")
```

Observe que, nessa implementação, a operação dequeue() possui complexidade 𝑂(1), pois envolve apenas a atualização da referência head. Entretanto, a operação enqueue() possui complexidade 𝑂(𝑛), uma vez que é necessário percorrer toda a lista até alcançar o último nó para inserir o novo elemento.
Essa limitação pode ser eliminada mantendo-se explicitamente um ponteiro adicional para a cauda da fila, permitindo que tanto enqueue() quanto dequeue() sejam executadas em tempo 𝑂(1). É possível fazer com que a operação enqueue tenha complexidade 𝑂(1). Para isso, devemos alterar a definição do TAD Dynamic_Queue, de modo a incluir uma referência explícita para o último nó
da lista (cauda).

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
# Inserir na filha
# Remover na fila
# Observar 

class Queue:
    
    def __init__(self):
        self.first = None
        self.last = None
        self._size = 0

    def push(self, elem):
        # Insere um elemento na fila
        node = Node(elem)
        if self.last is None:           # Fila vazia
            self.last = node            # Último nó da lista aponta para o novo nó 
            self.first = node           # Primeiro nó da lista aponta para o novo nó             
        else:
            self.last.next = node       # Último elemento aponta para o novo nó
            self.last = node            # Novo nó passa a ser o último elemento da fila
        self._size = self._size + 1

    def pop(self):
        # remove um elemento do ínicio da fila
        if self._size > 0:
            elem  = self.first.data
            self.first = self.first.next
            self._size = self._size - 1
            return elem
        raise IndexError("The queue is empty")
        
    def peek(self):
        # retorna o início da fila
        if self._size > 0:
            elem = self.first.data
            return elem
        raise IndexError("The queue is empty")

    def __len__(self):
        # retorna o tamanho da fila
        return self._size
    
    def __repr__(self):
        if self._size > 0:
            r = ""
            pointer = self.first
            while(pointer):
                r = r + str(pointer.data) + " "
                pointer = pointer.next
            return r
        return "Empty Queue"
    
    def __str__(self):
        return self.__repr__()
```
