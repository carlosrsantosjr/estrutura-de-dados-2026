# Filas (FIFO)

Uma fila (queue) é uma estrutura de dados linear em que a inserção de elementos é realizada em
uma extremidade (final) e a remoção é realizada na outra extremidade (início). Assim como em uma fila de pessoas no mundo real, o primeiro a entrar será o primeiro a sair, o que caracteriza o princípio de ordenação FIFO (First In, First Out). Além disso, filas são estruturas restritivas no sentido de que um elemento não pode ultrapassar seu antecessor. A ordem de atendimento é estritamente determinada pela sequência de chegada.

Um exemplo clássico do uso de filas é o de um servidor de impressão. Esse servidor precisa lidar com múltiplas requisições de impressão que chegam ao longo do tempo. Para organizar essas requisições, o software da impressora mantém uma fila de impressão, garantindo que os trabalhos sejam processados sequencialmente, de acordo com a ordem de chegada.

## Implementação de um TAD Fila

Para implementar um Tipo Abstrato de Dados (TAD) Fila, devemos definir:
 - Suas variáveis internas (estrutura de armazenamento);
 - As operações permitidas sobre seus elementos.

As operações fundamentais são:

 - *enqueue(x)*: insere o elemento x no final da fila;
 - *dequeue()*: remove e retorna o elemento no início da fila;
 - *front()*: retorna o elemento no início sem removê-lo;
 - *is_empty()*: verifica se a fila está vazia.

Observe que, embora conceitualmente diferente, a implementação de uma fila possui operações
semelhantes às de uma pilha, diferenciando-se principalmente na política de inserção e remoção de elementos.

Durante a criação da fila, devemos inicializar a referência *head* com *None*, indicando que a fila está vazia, e definir o tamanho como zero.

```python
class Node:
    def __init__(self, data=None):
        self.data = data            # armazena o dado
        self.next = None            # referência para o próximo nó 
```

```python
class Queue:
    def __init__(self):
        self.head = None            # indica a posição para remoção (início da fila);
        self.tail = None            # indica a posição para inserção (final da fila);
        self._size = 0              # armazena o número atual de elementos na fila.
```

A operação *enqueue()* insere um novo elemento no final do encadeamento, isto é, na cauda da fila.
Observe que, nessa implementação, a operação *enqueue()* possui complexidade *𝑂(1)*, pois envolve apenas a atualização da referência *tail*. Essa referência elimina a necessidade de percorrer toda a lista até alcançar o último nó para inserir o novo elemento.

```python
# Insere um elemento na fila
def enqueue(self, elem):
        node = Node(elem)
        if self.tail is None:           # Fila vazia
            self.tail = node            # Último nó da lista aponta para o novo nó 
            self.head = node            # Primeiro nó da lista aponta para o novo nó             
        else:
            self.tail.next = node       # Último elemento aponta para o novo nó
            self.tail = node            # Novo nó passa a ser o último elemento da fila
        self._size = self._size + 1
```

A operação *dequeue()* remove o elemento presente no início da fila:

```python
# Remove um elemento do ínicio da fila
def dequeue(self):
    if self._size > 0:
        elem  = self.head.data
        self.first = self.head.next
        self._size = self._size - 1
        return elem
    raise IndexError("The queue is empty")
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data            # armazena o dado
        self.next = None            # referência para o próximo nó  

class Queue:
    def __init__(self):
        self.head = None            # indica a posição para remoção (início da fila);
        self.tail = None            # indica a posição para inserção (final da fila);
        self._size = 0              # armazena o número atual de elementos na fila.

    # Insere um elemento na fila
    def enqueue(self, elem):
        node = Node(elem)
        if self.tail is None:           # Fila vazia
            self.tail = node            # Último nó da lista aponta para o novo nó 
            self.head = node            # Primeiro nó da lista aponta para o novo nó             
        else:
            self.tail.next = node       # Último elemento aponta para o novo nó
            self.tail = node            # Novo nó passa a ser o último elemento da fila
        self._size = self._size + 1

    # Remove um elemento do ínicio da fila
    def dequeue(self):
        if self._size > 0:
            elem  = self.head.data
            self.head = self.head.next
            self._size = self._size - 1
            return elem
        raise IndexError("The queue is empty")

    # retorna o início da fila
    def front(self):
        if self._size > 0:
            elem = self.head.data
            return elem
        raise IndexError("The queue is empty")

    # retorna o tamanho da fila
    def __len__(self):
        return self._size
    
    # retorna itens da fila
    def show(self):
        pointer = self.head
        while pointer:
            print(pointer.data)
            pointer = pointer.next
```
